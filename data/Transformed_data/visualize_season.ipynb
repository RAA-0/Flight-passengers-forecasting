```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("Transformed_arrival.csv")

# Group data by month and calculate the mean or sum of passengers per month
monthly_seasonality = df.groupby('date_time_month')['total'].mean()

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_seasonality.index, y=monthly_seasonality.values)
plt.title('Average Passengers by Month')
plt.xlabel('Month')
plt.ylabel('Average Passengers')
plt.show()
```

```python
# Filter data for months 8 (August) and 12 (December)
specific_months = df[df['date_time_month'].isin([ 7,8, 12])]

# Group by month and day, and calculate the mean or sum of passengers per day
daily_passenger_counts = specific_months.groupby(['date_time_month', 'date_time_day_of_month'])['total'].mean()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(x=daily_passenger_counts.index.get_level_values('date_time_day_of_month'), 
             y=daily_passenger_counts.values, 
             hue=daily_passenger_counts.index.get_level_values('date_time_month').map({7:"July",8: 'August', 12: 'December'}))
plt.title('Daily Passenger Counts in August and December')
plt.xlabel('Day of the Month')
plt.ylabel('Average Passengers')
plt.show()
```
